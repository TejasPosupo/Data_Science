#### 7. Data Preprocessing

```python
movies.iloc[0].genres
```

* converting 1 --> 2

1. : '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
2.  : ['Action','Adventure','Fantasy','SciFi']

```python
def convert(dic):
    l = list()
    for i in ast.literal_eval(dic):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['genres']
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

* cast : We need only top 3 characters

```python
def convert_cast(dic):
    l = list()
    counter = 0
    for i in ast.literal_eval(dic):
        if counter != 3 :
            l.append(i['name'])
            counter += 1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

* We need only Director name

```python
def fetch_director(dic):
    l = list()
    for i in ast.literal_eval(dic):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies['crew']
```

```python
movies.head()
```

```python
movies.info()
```

* before merging columns we need to convert overview columns from <br>     string --> list form

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['overview'][0]
```

```python
movies.head()
```

Now we have to remove space b/w names because model will read it like two tags : <br>

* ex : 'Shivam Pal ---> 'ShivamPal'

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

 * Now we need to concanate 'overview','genres','keywords','cast','crew' columns in a single column as a tags

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['overview'] + movies['cast'] + movies['overview'] + movies['crew']
```

```python
movies.head()
```

* Now we select only 'id', 'original_title', 'tags', release_date' columns and make a new dataset

```python
df = movies[['id','original_title','tags','release_date']]
df
```

* Now we convert list into string

```python
df['tags'] = df['tags'].apply(lambda x:" ".join(x))
df
```

now converting all strings into lowercase for better performance and understanding

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df
```

#### 8. Now we perform text vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
vectors
```

```python
cv.get_feature_names()  #these are the frequent words
```

```python
len(cv.get_feature_names())
```

#### 8. Applying steming technique for converting similar words in only one words.
* Ex : ['apply','appling','applied'] --> ['apply','apply','apply']

```python
# !pip install nltk
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    a = list()
    for i in text.split():
        a.append(ps.stem(i))
    return " ".join(a)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
df['tags']
```

#### 9. Now Calculating distance b/w movies in vector space (through angle)

```python
from sklearn.metrics.pairwise import  cosine_similarity
```

```python
total_similarity = cosine_similarity(vectors)
total_similarity.shape
```

#### 10. Making a function for recommend 5 movies

```python
def recommend(movie):
    movie_index = df[df['original_title'] == movie].index[0]
    distances = total_similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list :
        print(df.iloc[i[0]].original_title)
```

```python
recommend('Avatar')
```
